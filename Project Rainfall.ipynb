```python
# import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# import and read the zipfile
import zipfile
zip = zipfile.ZipFile(r"F:Data Science\data science.zip")
zip.namelist()
```

```python
# read the dataset from csv
data = pd.read_csv(zip.open('data science/austin_weather.csv'))
data
```

```python
# replacing 'T' and '-' with '0.0'
data = data.replace('T', 0.0)
data = data.replace('-', 0.0)
data
```

```python
# deleting unnecessary columns
df = data.drop (["Date", "Events", "SeaLevelPressureLowInches"], axis= 1, inplace= True)
data
```

```python
# saving this modified csv into a new csv file
data.to_csv("austin_weather_final.csv")
```

```python
df = pd.read_csv("austin_weather_final.csv")
df
```

```python
# creating labels for the prediction model, attempting to keep PrecipitationSumInches as y
X = df.drop(["PrecipitationSumInches"], axis= 1)
Y = df["PrecipitationSumInches"]
```

```python
# reshaping Y into 2D vector
Y = Y.values.reshape(-1,1)
```

```python
# take any random day by indexing it
day_index = 607
days = [i for i in range(Y.size)]
```

```python
# initialize the linear regression classifier
clf = LinearRegression()
```

```python
# train the classifier
clf.fit(X,Y)
```

```python
# plot in graph
print ("The precipitation trend graph:")
plt.scatter(days, Y, color = 'blue')
plt.scatter(days[day_index], Y[day_index], color = 'red')
plt.title("Precipitation Level")
plt.xlabel("days")
plt.ylabel("Precipitation in Inches")
plt.show()
```

```python
x_vis = X.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
# to visualize precipitation trend with different attributes
print ("The precipitation vs the attributes trend graph:")
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color = 'blue')
    plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index], color = 'red') 
    plt.title(x_vis.columns.values[i])
plt.show()
```
